<h2 style="text-align:center;font-family:array;font-size:35px">La dichotomie </h2>

<p>Tout d'abord c'est quoi la dichotomie ? Ce mot vient du grecque pour dire "division en deux" et c'est par ce principe de couper en deux que nous arrivons a la méthode dichotomique</br>
La méthode dichotomique à pour but de trouver en quel point une fonction s'annule en coupant la fonction en deux petit a petit, ce point s'appel la racine, prenons l'exemple d'en dessous et observer la droite.
</p>

```python
#Nombre de point
n=100

#Intervalle
a=-10
b=10

#Abscisses
X=[a+(b-a)/n*k for k in range(n+1)]
X1=[x for x in X if(x>0)]
X2=[x for x in X if(x<0)]
#Ordonnées
Y1=[x+3-x*2 for x in X1]
Y2=[x+3-x*2 for x in X2]


#Axes du repère
plot([a, b], [0, 0], 'r', lw=1)
plot([0, 0], [min(Y2), max(Y1)], 'r', lw=1)

#Hyperbole
plot(X1, Y1, 'b', lw=2)
plot(X2, Y2, 'b', lw=2)#Titre
title("Une droite")#Affichage
show()
```

<p> Tout d'abord, on se place dans<strong> le domaine des réels</strong> donc des nombres entre <strong>moins l'infinie</strong> et <strong>plus l'infinie</strong>. Dans notre exemple la droite est décroissante et linéaire, observer lorsque la droite coupe <strong>l'axe des abscisses</strong>, on peut voir que <strong>le point 3 en abscisse</strong>, environ, coupe ce point, on en conclue alors que le point 3 est racine de notre fonction. </br>Cependant il est possible que certaines fonction n'est pas de racines réel :
<ul><li>Si f(a) > 0 et f(b) > 0 : La fonction n'a <strong>pas de solution</strong> donc <strong>pas de racines</strong></li>
<li>Si f(a) > 0 et f(b) < 0 :  La fonction a au moins <strong>une solution</strong> donc au moins <strong>une racine</strong></li>
    <li>Si f(a) < 0 et f(b) > 0 :  La fonction a au moins <strong>une solution</strong> donc au moins <strong>une racine</strong></li>
        <li>Si f(a) < 0 et f(b) < 0 : La fonction n'a <strong>pas de solution</strong> donc <strong>pas de racines</strong></li>
</ul>
</p>

<h2 style="text-align:center">Méthode dichotomique</h2>

<p>Dès lors que nous savons qu'il existe une racine nous pouvons commencer la méthode dichotomique:</br>
Dans cette méthode nous allons couper en deux des intervalles jusqu'a trouver notre racine avec une certaine précision : 
<ul>Pour trouver de quel côté commencer on prend le millieu et on l'utilise dans notre fonction , alors deux possibilité apparaisse</br><li>Si f(a+b/2)*f(a) > 0 : alors la racine se trouve dans l'intervalle [a+b/2;b] </li>
<li>Si f(a+b/2)*f(a) < 0 : alors la racine se trouve dans l'intervalle [a;a+b/2] </li>
</ul></br>
    On va répéter ce principe jusqu'a avoir un intervall avec la précision qu'on souhaite , on peut donc transcrire ceci dans suite en python 
</p>

```python
millieu = (a+b)/2
if f(millieu)*f(a)> 0:
    a=millieu
else:           
    b=millieu
```

<p>Il nous faut maintenant savoir quand s'arrête notre boucle, et pour faire cela on va utiliser un <strong>while</strong>.<br>
Dans notre fonction on va utiliser une valeur pour savoir a combien de chiffre après la virgule on veut que notre racine soit trouver on va l'appeler k , si k = 0.01 alors tant que  la valeur absolute de a-b est plus petit que k on va continuer notre méthode :
<ul><li>Tant que |a-b| > k alors on continue</li></ul>

En plus d'avoir notre boucle il nous faut notre racine et on sait que si f(millieu) est égale a 0 alors c'est que la valeur milleu est égale au réel de notre racine :
<ul><li>Si f(a+b/2)=0 alors (a+b)/2 = racine</li></ul>
Et enfin faisons en sorte de prendre une fonction avec le bonne encadrement pour éviter de se retrouver sans racine, transformons ceci en un test :
<ul><li>Si f(a)*f(b) > 0 : alors le choix de l'encadrement est mauvais </ul>
</p>

```python
if f(a)*f(b)>0:
    print('Mauvais encadrement !')
k=0.01
while abs(a-b) > k:
    millieu = (a+b)/2
    if f(millieu) == 0:
        print(millieu+" est notre racine")
    if f(millieu) > 0:
        a=millieu
    else:
        b=millieu
```

<p>Nous avons élaborer en python toutes les étapes de notre méthode , finissons donc en la mettant dans une fonction.<br>
En paramètre de notre fonction mettons :
    <ul><li>Notre fonction f</li>
<li>Notre valeur a</li>
<li>Notre valeur b</li>
<li>Notre valeur k pour définir la précision</li>
</ul>
</p>

```python
def dichotomie(f,a,b,k):
    if f(a)*f(b)>0:
        print('Mauvais encadrement !')
        return 
    millieu = (a+b)/2
    while abs(a-b) > k:
        if f(millieu) == 0:
            return millieu
        if f(millieu)*f(a)> 0:
            a=millieu
        else:
            b=millieu
        millieu = (a+b)/2
    return millieu
```

<p>Notre fonction est prête utilisons des fonctions pour les tester ! <br>
<ul><li>Première fonction : x+3-x*2</li> 
<li>Deuxième fonction : x**2+x*3-1</li>
<ul></p>

```python
#Nombre de point
n=100

#Intervalle
a=-10
b=10

#Abscisses
X=[a+(b-a)/n*k for k in range(n+1)]
X1=[x for x in X if(x>0)]
X2=[x for x in X if(x<0)]

#Ordonnées
Y1=[x+3-x*2 for x in X1]
Y2=[x+3-x*2 for x in X2]

#Quadrillage
grid(True)

#Axes du repère
plot([a, b], [0, 0], 'r', lw=1)
plot([0, 0], [min(Y2), max(Y1)], 'r', lw=1)

#Hyperbole
plot(X1, Y1, 'b', lw=2)
plot(X2, Y2, 'b', lw=2)#Titre
title("Une droite")#Affichage
show()
```

<p>Sur cette droite on obverse que la droite coupe la ligne des abscisses entre 0 et 5 , alors on va prendre un intervalle de 
[0;5] en plus de cela prenons une valeur k pour notre précision de 0.1
</p>

```python
def f(x):
    return x+3-x*2

print(dichotomie(f,0,5,0.1))
```

<p>Bien nous avons trouvé notre racine ! Elle est de 3 environ </p>

```python
#Nombre de point
n=100

#Intervalle
a=-10
b=10

#Abscisses
X=[a+(b-a)/n*k for k in range(n+1)]
X1=[x for x in X if(x>0)]
X2=[x for x in X if(x<0)]

#Ordonnées
Y1=[ x**2+x*3-1 for x in X1]
Y2=[ x**2+x*3-1 for x in X2]

#Quadrillage
grid(True)

#Axes du repère
plot([a, b], [0, 0], 'r', lw=1)
plot([0, 0], [min(Y2), max(Y1)], 'r', lw=1)

#Hyperbole
plot(X1, Y1, 'b', lw=2)
plot(X2, Y2, 'b', lw=2)#Titre
title("Une fonction")#Affichage
show()
```

<p>Pour cette deuxième fonction on peut observer qu'il y a deux racines en utilisent un bon encadrement nous pouvons avoir les deux racines, en prenant un interval de -5 a -2 on a une racine puis avec un interval de -1 a 2 on a la deuxième racine</p>

```python
def f(x):
    return x**2+x*3-1

#Première racine dans l'interval [-5;-1]
premiere=dichotomie(f,-5,-1,0.1)
print(premiere)

#Deuxième racine dans l'interval [-1;5]
deuxieme=dichotomie(f,-1,5,0.1)
print(deuxieme)
```

<p>Grace a notre bon encadrement nous avons eu nos deux racines qui sont : <ul><li>Interval [-5;-1] : -3.3</li> <li>Interval [-1;5] : 0.27</li><ul></p>

<h2 style="text-align:center">Consommation de la méthode dichotomique </h2>

<p>Parlons a présent du bénéfice de la consommation en temps et en mémoire de notre méthode !Pour ce faire nous allons la comparer a d'autre méthode de recherche. Faisons une comparaison avec la recherche séquentielle.Voici ci-dessous en python la recherche séquentielle on va comparer son temps de recherche face a celui de la dichotomie pour trouver une valeur dans un tableau </p>

```python
def recherche_seq(element, tab):
    i = 0
    while i < len(tab) and tab[i] != element:
        i += 1
    if i == len(tab):
        return -1  
    else:
        return i

def recherche_dt(element, tab):
    debut = 0
    fin = len(tab) - 1
    while debut <= fin:
        milieu = (debut + fin) // 2

        if element == tab[milieu]:
            return milieu
        elif element < tab[milieu]:
            fin = milieu - 1  
        else:
            debut = milieu + 1  
    return -1

```

<p>Juste au dessus on réecris notre fonction dichotomique pour qu'elle cherche dans un tableau si un nombre existe. C'est exactement la même méthode que pour trouver nos racines !<br> Et de plus on écrit une fonction de recherche séquentielle</p>

```python
# Fonction qui nous permet de créer un tableau 
def cree_tab(n):
    t = []
    i = 0
    while i < n:
        t.append(i)
        i += 1
    return t

# Dans cette fonction on va faire nos calcul sur plusieurs mesures(20) , et on va calculer le temps de chaque mesure 
# pour la recherche séquentielle et la méthode dichotomique puis on va les comparer 

def mesure_temps_recherche(n, nb_mesures):
    tps = [0., 0.]
    tableau = cree_tab(n)
    i = 0
    while i < nb_mesures:
        element = n
        tic = time()
        res = recherche_seq(element, tableau)
        tps[0] += time() - tic

        tic = time()
        res = recherche_dt(element, tableau)
        tps[1] += time() - tic

        i += 1

    tps[0] = round(1000 * tps[0] / nb_mesures, 6)
    tps[1] = round(1000 * tps[1] / nb_mesures, 6)
    return tps


nb_mesures = 20
tailles = [100, 500, 1000, 5000, 10000, 50000, 100000, 500000, 1000000]
temps_seq = []
temps_dt = []
i = 0
while i < len(tailles):
    tps = mesure_temps_recherche(tailles[i], nb_mesures)
    temps_seq.append(tps[0])
    temps_dt.append(tps[1])
    i += 1
```

<p>Grâce au script en python juste au-dessus nous allons comparer le temps de recherce des deux algorithmes</p>

```python
print("Temps moyen de recherche :")
i = 0
while i < len(tailles):
    print("Pour un tableau de taille ", tailles[i], ":")
    print("\t", temps_seq[i], "ms pour la recherche séquentielle")
    print("\t", temps_dt[i], "ms pour la recherche dichotomique")
    i += 1
```

<p>En comparaison a la recherche séquentielle on voit que la méthode dichotomique prend moins de 0.0 ms alors que pour la recherche séquentielle elle en prend de plus en plus avec la taille de l'échantillon qui augmente<br>
Mettons ceci dans un graphe pour mieux comprendre !</p>

```python
%matplotlib inline

plot(tailles, temps_seq, label="séquentiel")
plot(tailles, temps_dt, label="dichotomie")
xlabel("taille")
ylabel("temps (ms)")
legend()
```

<p>On observe donc que la recherche dichotomique est très largement supérieur a la recherche séquentielle et ce sur plusieurs échantillons allant à 400, alors que en contre partie la recherche séquentielle ne fait qu'augmenter linéairement.  </p>
